```python
import pandas as pd
import os
from dotenv import load_dotenv
from langchain.chains.question_answering import load_qa_chain
from langchain_openai import OpenAI
```

```python
load_dotenv()
OPENAI_KEY=os.environ["OPENAI_API"]
```

```python
df = pd.read_csv('../test-dataset/testset.csv')
df.head()
```

```python
chain = load_qa_chain(
    OpenAI(openai_api_key=OPENAI_KEY, temperature=0),
    chain_type="stuff")

def generate_answer(question):
    answer = chain.run(input_documents=[], question=question)
    return answer

answers = []
for question in df['question']:
    answer = generate_answer(question)
    answers.append(answer)

df['Answer'] = answers
df.to_csv('../test-dataset/dataset_with_answers.csv', index=False)
```
