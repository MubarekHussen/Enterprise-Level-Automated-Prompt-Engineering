```python
from dotenv import load_dotenv
import os

load_dotenv()

OPENAI_KEY=os.environ["OPENAI_API"]
```

```python
from langchain.document_loaders import DirectoryLoader
from ragas.testset.generator import TestsetGenerator
from ragas.testset.evolutions import simple, reasoning, multi_context

loader = DirectoryLoader('../pdfs', glob="**/*.pdf")
documents = loader.load()

generator = TestsetGenerator.with_openai(critic_llm = "gpt-4")

distributions = {
    simple: 0.5,
    multi_context: 0.4,
    reasoning: 0.1
}

testset = generator.generate_with_langchain_docs(documents, 10, distributions)
testset.to_pandas()
```

```python
test_df = testset.to_pandas()
```

```python
test_df.to_csv('../test-dataset/testset.csv', index=False)
```
